#### Feature Selection

```python
X = df.iloc[:,2:5]
Y = df.iloc[:,5:6]

print('Data X/Independen\n',X)
print('Data Y/Dependen\n',Y)
```

#### Normalisasi

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

scaled_X = min_max_scaler.fit_transform(X)
scaled_Y = min_max_scaler.fit_transform(Y)

print('Data X setelah dinormalisasi\n',scaled_X)
print('Data Y setelah dinormalisasi\n',scaled_Y)
```

```python
scaled_X = np.array(scaled_X)
scaled_Y = np.array(scaled_Y)
```

#### Split DTr DTs

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(scaled_X,scaled_Y,test_size=0.12,random_state=40)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, criterion='squared_error',max_depth=10,min_samples_split=10,max_leaf_nodes=10,n_jobs=25,random_state=0,verbose=3)

rf.fit(X_train, Y_train)
```

```python
rf_predict = rf.predict(X_test)

rf_predict
```

#### Evaluasi Model RF

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

mae = mean_absolute_error(Y_test,rf_predict)
mse = mean_squared_error(Y_test,rf_predict)

print('Nilai MAE RF : ',mae)
print('Nilai MSE RF : ',mse)
```

```python
from math import sqrt

rmse = sqrt(mse)
print('Nilai RMSE RF: ',rmse)
```

```python
from sklearn.metrics import r2_score
print('Hasil R2 Score RF: ',r2_score(Y_test,rf_predict)*100)
```

```python
rf_predict_dua = rf.predict(scaled_X)

rf_predict_dua
```

```python
rf_predict_dua = rf_predict_dua.reshape(-1,1)
```

```python
rf_predict_dua
```

```python
inv_rf_predict_dua = min_max_scaler.inverse_transform(rf_predict_dua)

inv_rf_predict_dua
```

```python
inv_rf_predict_dua = inv_rf_predict_dua.reshape(-1)
```

```python
df_rf_predict_dua = pd.DataFrame({'Prediction RF':inv_rf_predict_dua})

df_rf_predict_dua
```

```python
df = df.merge(df_rf_predict_dua,how='left',left_index=True,right_index=True)

df
```

```python
df.drop(columns='Prediction RF_x',axis=1,inplace=True)
df.rename({'Prediction RF_y':'Prediction RF'},axis=1,inplace=True)
```

```python
df
```

```python
half_year_locator = mdates.MonthLocator(interval=6)
year_month_formatter = mdates.DateFormatter("%Y-%m")
month_year_formatter = mdates.DateFormatter('%b, %Y')

fig,ax = plt.subplots(figsize=(30,10))
ax.xaxis.set_major_locator(half_year_locator)
ax.xaxis.set_major_formatter(year_month_formatter)
ax.xaxis.set_major_formatter(month_year_formatter)

fig.autofmt_xdate()

# plt.figure(figsize=(15,7))
sns.set_theme(style='white',font_scale=2.1)


plt.plot(df['Date'],df['Close'],color='teal',label='Actual Close Price')
plt.plot(df['Date'],df['Prediction RF'],color='orange',label='Prediction Close Price using Random Forest')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.title('Comparison of Prediction & Actual Stock Price PT. XYZ 02/01/2020 - 12/02/2024, Tr 88% Ts 12%')
plt.show()
```

#### SVR RBF

```python
from sklearn.svm import SVR

model_svr = SVR(kernel='rbf', C=0.1, gamma=1)

model_svr.fit(X_train,Y_train)
```

#### Optimasi Grid Search CV

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1,1,10], 'gamma': [0.1,1, 10],'kernel': ['rbf','sigmoid']}

search = GridSearchCV(estimator=SVR(),param_grid=param_grid,refit=True, verbose=2)

search.fit(X_train,Y_train)
```
